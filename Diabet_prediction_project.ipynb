```python
from sklearn.preprocessing import StandardScaler

# Création d'un objet StandardScaler
scaler = StandardScaler()

# Sélection des colonnes à mettre à l'échelle
columns_to_scale = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction','Age']

# Ajuster le scaler aux données et transformer chaque colonne
for column in columns_to_scale:
    df[column] = scaler.fit_transform(df[[column]])



# Count the occurrences of each category
diabetic_count = df['Outcome'].sum()
non_diabetic_count = len(df) - diabetic_count

# Create labels for the pie chart
labels = ['Diabetic', 'Non-Diabetic']

# Count values for each label
sizes = [diabetic_count, non_diabetic_count]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the pie chart
plt.show()
```

```python
from sklearn.model_selection import train_test_split

target_column= 'Outcome'
X = df.drop(columns=[target_column])
y = df[target_column]

```

```python
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd


# Initialize the RandomUnderSampler
rus = RandomUnderSampler(random_state=42)

# Undersample the dataset
X_resampled, y_resampled = rus.fit_resample(X, y)

# Check the class distribution after undersampling
print(pd.Series(y_resampled).value_counts())

x_train, x_test, y_train, y_test = train_test_split(X_resampled, y_resampled, random_state=42, train_size=0.75)
# x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size=0.75)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import svm

model = svm.SVC(kernel='rbf', gamma='scale')
model.fit(x_train,y_train)

y_pred = model.predict(x_test)

score = accuracy_score(y_test,y_pred)
print(score)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Matrice de confusion :")

print(conf_matrix)
```
